```python
import matplotlib.pyplot as plt
```

```python
# Age groups
age_g1 = [1, 3, 5, 10, 15, 17, 18, 16, 19, 21, 23, 28, 30, 31, 33, 38, 32, 40, 45, 43, 49, 55, 53,
63, 66, 85, 80, 57, 75, 93, 95]
age_g2 = [6, 4, 15, 17, 19, 21, 28, 23, 31, 36, 39, 32, 50, 56, 59, 74, 79, 34, 98, 97, 95, 67,
69, 92, 45, 55, 77, 76, 85]

# Plotting with 10 bins
plt.hist(age_g1,bins=10,alpha=0.5, edgecolor='black',label='Age Group 1')
plt.hist(age_g2,bins=10,alpha=0.5, edgecolor='red',label='Age Group 2')

# Labels and legends
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Groups')
plt.legend()

# Show plot
plt.show()
```

# Ques-2

```python
import pandas as pd
```

```python
# Load the iris dataset
iris_data = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
```

```python
# Specify the features to plot
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

# Custom colors
colors = ['red', 'green', 'yellow', 'purple']

# Loop through each feature
for i, feature in enumerate(features):
    plt.hist(iris_data[feature], bins=20, alpha=0.7, color=colors[i], label=feature)

# Labels and legends
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histograms of Iris Features')
plt.legend()

# Show plot
plt.show()
```

# Ques-3

```python
# Arrays representing the number of students in IT, ECE, and CSE from years 2015-2019
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Years from 2015-2019
years = [2015, 2016, 2017, 2018, 2019]

# Width of each bar
bar_width = 0.25

# Plotting the IT bars
plt.bar(years, IT, bar_width, color='red', label='IT')

# Plotting the ECE bars
plt.bar([year + bar_width for year in years], ECE, bar_width, color='green', label='ECE')

# Plotting the CSE bars
plt.bar([year + 2*bar_width for year in years], CSE, bar_width, color='blue', label='CSE')

# Labels and legends
plt.xlabel('Year')
plt.ylabel('Number of Students')
plt.title('Number of Students Passed in Engineering Branch')
plt.legend()

# Show plot
plt.show()
```

# Ques-4

```python
import numpy as np

# Sample data
means_men = [22, 30, 35, 35, 26]
means_women = [25, 32, 30, 35, 29]
std_men = [4, 3, 4, 1, 5]
std_women = [3, 5, 2, 3, 3]

# Number of groups
N = 5

# Creating the bar plot
fig, ax = plt.subplots()

# Men bars
men_bar = ax.bar(np.arange(N), means_men, 0.35, yerr=std_men, color='red', label='Men')

# Women bars
women_bar = ax.bar(np.arange(N), means_women, 0.35, bottom=means_men, yerr=std_women, color='green', label='Women')

# Adding labels and title
ax.set_xlabel('Group')
ax.set_ylabel('Mean Values')
ax.set_title('Stacked Bar Plot with Error Bars')
ax.set_xticks(np.arange(N))
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
ax.legend()

# Show plot
plt.show()
```

# Ques-5

```python
# Sample data
year = [2014, 2015, 2016, 2017, 2018, 2019]
issues_addressed = [10, 14, 0, 10, 15, 15]
issues_pending = [5, 10, 50, 2, 0, 10]

# Creating the bar plot
fig, ax = plt.subplots()

# Addressed bars
addressed_bar = ax.barh(year, issues_addressed, color='red', label='Issues Addressed')

# Pending bars
pending_bar = ax.barh(year, issues_pending, left=issues_addressed, color='yellow', label='Issues Pending')

# Adding labels and title
ax.set_xlabel('Number of Issues')
ax.set_ylabel('Year')
ax.set_title('Horizontal Stacked Bar Chart')
ax.legend()

# Show plot
plt.show()
```
